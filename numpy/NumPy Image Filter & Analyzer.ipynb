```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=np.arange(0,25,dtype='uint8').reshape(5,5)*10
```

```python
brightened=image+50
flattened=image.ravel()
```

```python
column_total=image.sum(axis=0)
```

```python
print("Original Image Matrix:\n", image)
```

```python
print("\nVertical Brightness (Axis 0 Sum):\n", column_total)
```

```python
plt.figure(figsize=(8, 5))
plt.bar(range(5),column_total,color='gold',edgecolor='white')
plt.title("Vertical Brightness (Sum along Axis 0)")
plt.xlabel("Column Index")
plt.ylabel("Total Intensity Sum")
plt.grid(axis='y', linestyle='--', alpha=0.6)
```
