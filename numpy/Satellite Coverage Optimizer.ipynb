```python
import numpy as np
import matplotlib.pyplot as plt
```

coordinate initialized of satellite revolution

```python
cord=np.random.uniform(0,100,size=(50,2))
```

```python
user_pos=np.array([50,50])
```

use distance formula to find distance b/w user and satellite

```python
diff=user_pos - cord
```

```python
sq_diff = diff**2
```

```python
sum_sq= np.sum(sq_diff, axis=1)
```

```python
dist= np.sqrt(sum_sq)
```

```python
dist
```

```python
dist.shape
```

finding index position for the min distance of satellite from user_pos


```python
index_min = np.argmin(dist)
```

boolean masking for see which satellite are in range

```python
mask = dist <15
```

```python
dist[mask[mask<15]]
```

```python
len(dist[mask])
```

```python
# Data preparation
in_range = len(dist[mask])
out_range = len(dist) - in_range

plt.figure(figsize=(6, 6))
plt.pie([in_range, out_range], labels=['In Range', 'Out of Range'],
        autopct='%1.1f%%', colors=['#2ecc71', '#e74c3c'], startangle=90, explode=[0.1, 0])

plt.title("Satellite Coverage Success Rate")
plt.show()
```
