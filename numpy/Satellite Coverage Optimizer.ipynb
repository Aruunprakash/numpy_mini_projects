```python
import numpy as np
```

coordinate initialized of satellite revolution

```python
cord=np.random.uniform(0,100,size=(50,2))
```

```python
user_pos=np.array([50,50])
```

use distance formula to find distance b/w user and satellite

```python
diff=user_pos - cord
```

```python
sq_diff = diff**2
```

```python
sum_sq= np.sum(sq_diff, axis=1)
```

```python
dist= np.sqrt(sum_sq)
```

finding index position for the min distance of satellite from user_pos


```python
index_min = np.argmin(dist)
```
